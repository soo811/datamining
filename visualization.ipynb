```python
date = pd.read_csv('./data/autumn_day.csv')
# bukhan = date.iloc[:, 1].to_numpy()
# chiak = date.iloc[:, 2].to_numpy()
# gyeryong = date.iloc[:, 3].to_numpy()
# jiri = date.iloc[:, 4].to_numpy()
# naejang = date.iloc[:, 5].to_numpy()
year = date.iloc[:, 0].to_numpy()
```

```python
labels = ['Bukhan', 'Chiak', 'Gyeryong', 'Jiri', 'Naejang']
fig = plt.figure(figsize=(10,8))
for i, label in enumerate(labels):
    x = year
    y = date.iloc[:, i+1].to_numpy()
    sns.lineplot(x=x, y=y, label=label)
    plt.legend()
plt.xlabel('Year')
plt.ylabel('Starting Date')
```
